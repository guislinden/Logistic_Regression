```r
#Vamos aplicar o modelo de regressão logística na parte de Train da nossa dataset 
model <- glm(Class ~.,family=binomial(link='logit'),data=df)
summary(model)
```

```r
#Agora vamos remover as variáveis insignificantes (com menos de *)
df_limpo  <- df[,-c(4:8,13,14,18,19,21:25,27:30,32:37,39:42,44,45,48,49,50,51,52,54:62)]
```

```r
modelimpo <- glm(Class ~.,family=binomial(link='logit'),data=df_limpo)
summary(modelimpo)
```

```r
#Vamos verificar se existe multicolinearidade no nosso modelo. 
#Ou seja, verificar se no qual as variáveis independentes possuem relações lineares exatas ou aproximadamente exatas.
#Se o vif for > 2, é melhor deletar
vif(modelimpo)

```

```r
#Percebemos que OtherDebtorsGuarantors.None é multicolinear e OtherDebtorsGuarantors.CoApplicant quase é multicolinear. 
#Vamos deletá-las 
df_limpo$OtherDebtorsGuarantors.None <- NULL
df_limpo$OtherDebtorsGuarantors.CoApplicant  <- NULL 
```

### O modelo está criado!
Já possuimos as principais variáveis que geram um bom e mal devedor de crédito. Agora, vamos verificar a performance do modelo criado nos dados que reservamos para teste.
O ojetivo da avaliação de performance do modelo e validação é revisar:

- **Model Predictive Power** - Quão bom é o modelo na discriminação de casos de eventos e não eventos?

- **Generalização** - O modelo é generalizado entre as amostras, para que possa ser usado em qualquer amostra futura. Nós não queremos um modelo over-fitted (quando um modelo memoriza uma amostra de treinamento). Neste exemplo, desenvolveremos um modelo na amostra de treino e validaremos em outra amostra de teste para verificar se há excesso de ajuste.



#### Formas de metrificar o **Model Predictive Power**
- Confusion Matrix
- Concordance
- Gini Coefficient
- ROC /AUC
- Gains Table and Rank Ordering
- Lift Chart
- Cumulative Lift Chart
- KS Statistics


```r
#Vamos dividir entre treino e teste
df_limpo_treino  <- df_limpo[1:700,]
df_limpo_teste  <-  df_limpo[701:1000,]
```

```r
modelimpo_treino <- glm(Class ~.,family=binomial(link='logit'),data=df_limpo_treino)
summary(modelimpo_treino)
```

```r
df_limpo_teste$Pred_prob <- predict(modelimpo_treino,
                                  df_limpo_teste,
                                  type = c("response")
                                  )
```

```r
df_limpo_treino
```

```r
df_limpo_teste$Pred_Class <- ifelse(df_limpo_teste$Pred_prob>0.67,1,0)
```

```r
df_limpo_teste
```

```r
# Criar matris de confusão 
confusionMatrix(data=factor(df_limpo_teste$Pred_Class),
                reference=df_limpo_teste$Class,
                positive='1')

#verificamos accuracy de aproximadamente 75% na nossa amostra de teste. Ou seja, acertamos 75% se o credor é "Bad" ou "Good"
```

```r
#curva ROC: curva entre Falso Positivo e Verdadeiro Positivo 

library(ROCR)
 
perf.obj <- prediction(predictions=df_limpo_teste$Pred_prob,
                       labels=df_limpo_teste$Class)
# informações para o gráfico. 
roc.obj <- performance(perf.obj, measure="tpr", x.measure="fpr")
plot(roc.obj,
     main="German Credit - ROC Curves",
     xlab="1 - Specificity: False Positive Rate",
     ylab="Sensitivity: True Positive Rate",
     col="blue")
abline(0,1,col="grey")
```
